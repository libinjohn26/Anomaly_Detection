Calculating reference value for all data including anomaly data(Which is class 9 in this case)

```
max_sm_all_wt = []
for data, label in zip(x_test, y_test):
  img = data.reshape([-1, 28, 28, 1])
  ref_value = []
  
  if label == 9:
    lbl = ood
  else:
    lbl = ind

  max0 = model0.get_max_pred_value(img, temp=temp_value[0])
  max1 = model1.get_max_pred_value(img, temp=temp_value[1])
  max2 = model2.get_max_pred_value(img, temp=temp_value[2])
  max3 = model3.get_max_pred_value(img, temp=temp_value[3])
  max4 = model4.get_max_pred_value(img, temp=temp_value[4])
  max5 = model5.get_max_pred_value(img, temp=temp_value[5])
  max6 = model6.get_max_pred_value(img, temp=temp_value[6])
  max7 = model7.get_max_pred_value(img, temp=temp_value[7])
  max8 = model8.get_max_pred_value(img, temp=temp_value[8])
  ref_value =[max0, max1, max2, max3, max4, max5, max6, max7, max8]
  max_sm_all_wt.append([ref_value, lbl])
```

```
accuracy = 0.0
count_in = 0
count_ood = 0
acc_in = 0.0
acc_ood = 0.0

for data, lbl in max_sm_all_wt:
  dist_ood = np.linalg.norm(np.array(data) - np.array(ref_vec_wt_ood)) 
  dist_in = np.linalg.norm(np.array(data) - np.array(ref_vec_wt_in))

  sim_in = 1 / (1 + dist_in)
  sim_ood = 1 / (1 + dist_ood)

  #if dist_ood < dist_in:
  if lbl == ind:
    count_in = count_in + 1
    if sim_in > sim_ood:
      accuracy = accuracy + 1
      acc_in = acc_in + 1
  else:  
    count_ood = count_ood + 1
    if sim_ood > sim_in:
      accuracy = accuracy + 1
      acc_ood = acc_ood + 1

print("Total Accuracy: ", accuracy/len(max_sm_all_wt)) 
print("Accuracy of determining ID data: ", acc_in/count_in)
print("Accuracy of determining OOD data: ", acc_ood/count_ood)
```
